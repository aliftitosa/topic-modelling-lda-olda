```python
from tweepy.streaming import StreamListener
from tweepy import OAuthHandler
from tweepy import Stream
import json
from datetime import datetime, timedelta

access_token = "**************"
access_token_secret = "**************"
consumer_key = "**************"
consumer_secret = "**************"

class StdOutListener(StreamListener):
    start_time = datetime.now()
    index = 0
    delta = [8, 3, 3]
    
    def on_data(self, data):
        realtime = datetime.now()
        filename = "tweet_pemerintah_" + str(self.start_time.day) + "_" + str(self.start_time.month) + ".txt"

        if (realtime - self.start_time).days == self.delta[self.index]:
            self.index += 1
            self.start_time = realtime
            
        with open(filename, "a", encoding="utf-8") as tweet_log:
            tweet_log.write(data)
        

    def on_error(self, status):
        print (status)

l = StdOutListener()
auth = OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
stream = Stream(auth, l, tweet_mode='extended')
stream.filter(track=["pemerintah"])
```

```python
import json
import pandas as pd
import nltk
from nltk.tokenize import TweetTokenizer as tw_tokenizer
from unidecode import unidecode
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from nltk.tag import CRFTagger
from spacy.lang.id import Indonesian
import taudataNlpTm as tau, itertools, re, pickle, pyLDAvis, pyLDAvis.sklearn, spacy, urllib.request
import time, numpy as np, matplotlib.pyplot as plt, networkx as nx, seaborn as sns 
from tqdm import tqdm
from textblob import TextBlob
from pprint import pprint


#Gensim
import gensim
import gensim.corpora as corpora
from gensim.utils import simple_preprocess
from gensim.models import CoherenceModel
from gensim.models.ldamodel import LdaModel
from gensim.models import Phrases

#Visualazation
import pyLDAvis
import pyLDAvis.gensim
```

```python
class Preprocessing:
    
    def __init__(self):
        self.id_stop = self.get_stopword()
        self.slangS = self.get_slangs()
        self.stemmer = StemmerFactory().create_stemmer()
        self.nlp_id = Indonesian()  # Language Model
        self.ct = CRFTagger()
        self.ct.set_model_file('data/all_indo_man_tag_corpus_model.crf.tagger')
    
    def filtering(self, tweet):
        text = unidecode(tweet).lower()
        tokenized = tw_tokenizer().tokenize(text)
        filtered = [word for word in tokenized if word.isalnum()]
            
        for i, word in enumerate(filtered):
            if word in self.slangS.keys():
                filtered[i] = self.slangS[word]

        remove_stopword = [self.stemmer.stem(word) for word in filtered if word not in self.Sastrawi_StopWords_id] 

        return ' '.join(remove_stopword)
    
    def get_stopword(self):
       
        factory = StopWordRemoverFactory()

        self.Sastrawi_StopWords_id = factory.get_stop_words()
        #load stopword yang ada di modul

        with open('data/stopwords_id.txt', "r", encoding="utf-8", errors='replace') as f:
            id_stop = f.readlines()

        return [t.strip().lower() for t in id_stop]
        
    def get_slangs(self):
        
        with open('data/slang.txt', "r", encoding="utf-8", errors='replace') as f:
            slangS = f.readlines()
            
        slangS = [t.strip('\n').strip().split(':') for t in slangS]
        return {k.strip(): v.strip() for k, v in slangS}
    
    def PosTags(self, t, filters):
        tokens = self.nlp_id(self.filtering(t))
        tokens = [str(k) for k in tokens if len(k)>2]
        hasil = self.ct.tag_sents([tokens])
        return [k[0] for k in hasil[0] if k[1] in filters]
        
```
